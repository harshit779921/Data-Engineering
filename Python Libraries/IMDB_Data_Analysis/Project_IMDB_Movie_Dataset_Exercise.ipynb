### 20. Classify Movies Based on Ratings [Excellent, Good and Average]

```python
df.columns
```

```python
def rating(rating):
    if rating >= 7.0:
        return "Excellent"
    elif rating >=6.0:
        return "Good"
    elif rating >= 5.0:
        return "Average"
```

```python
df["rating_cat"] = df["Rating"].apply(rating)
```

### 22. Find Unique Values From Genre 

```python
df.columns
```

```python
df["Genre"]
```

```python
list1=[]
for value in df["Genre"]:
    list1.append(value.split(","))
```

```python
list1 
```

```python
One_D=[]
for item in list1:
    for item1 in item:
        One_D.append(item1)
```

```python
One_D
```

```python
uni_list=[]
for item in One_D:
    if item not in uni_list:
        uni_list.append(item)
        
```

```python
uni_list
```
